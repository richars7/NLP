```python
import bs4 as bs
```

```python
import urllib.request
import re
import nltk
nltk.download('stopwords')
import heapq
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Latent_semantic_analysis').read()
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
text=""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
clean_text = text.lower()
clean_text = re.sub(r'\W',' ',clean_text)
clean_text = re.sub(r'\d',' ',clean_text)
clean_text = re.sub(r'\s+',' ',clean_text)
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

# Word counts 

```python
word2count = {}
for word in nltk.word_tokenize(clean_text):
    if word not in stop_words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1
            
```

```python
word2count
```

```python
for key in word2count.keys():
    word2count[key] = word2count[key]/max(word2count.values())
```

# Sentence Scores

```python
sent2score ={}
for sent in sentences:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word2count.keys():
            if len(sent.split(' '))< 25:
                if sent not in sent2score.keys():
                    sent2score[sent] = word2count[word]
                else:
                    sent2score[sent]+= word2count[word]
```

```python
sent2score
```

```python
best_sent = heapq.nlargest(5, sent2score, key = sent2score.get)
```

```python
for sent in best_sent:
    print(sent)
```
